### Lambda Functions

Lambda function adalah fungsi anonim (tanpa nama) di Python.

lambda argumen: ekspresi

```python
kali = lambda a, b : a * b
print(kali(3,2))
```

```python
n = lambda x: "Positif" if x > 0 else "Negatif" if x < 0 else "Zero"

print(n(5))
print(n(-3))
print(n(0))
```

```python
check = lambda x: "Genap" if x % 2 == 0 else "Ganjil"

print(check(4))
print(check(7))
```

```python
li = [lambda arg=x: arg * 10 for x in range(1,5)]
for i in li:
    print(i())
```

```python
n = [1,2,3,4,5,6]
even = filter(lambda x: x % 2 == 0, n)
print(list(even))
```

```python
a = [1,2,3,4]
b = map(lambda x: x * 2, a)
print(list(b))
```

```python
from functools import reduce
a = [1,2,3,4,5]
b = reduce(lambda x, y: x * y, a)
print(b)
```

Perbedaan keyword lambda dan def

```python
# Lambda
sq = lambda x: x ** 2
print(sq(2))

# Def
def sqdef(x):
    return x ** 2
print(sqdef(2))
```

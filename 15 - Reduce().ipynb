### reduce()

reduce() adalah fungsi dari modul functools yang digunakan untuk menggabungkan elemen-elemen dalam iterable menjadi satu nilai dengan cara menerapkan sebuah fungsi secara berulang.

from functools import reduce

reduce(function, iterable[, initializer])

* function → fungsi yang menerima dua argumen dan mengembalikan satu nilai.
* iterable → kumpulan data (list, tuple, dll).
* initializer (opsional) → nilai awal sebelum proses reduksi dimulai.

```python
from functools import reduce

angka = [1,2,3,4,5]

hasil = reduce(lambda a,b: a * b, angka)
print(hasil)
```

```python
def add(x,y):
    return x + y

a = [1,2,3,4,5]
res = reduce(add,a)
print(res)
```

```python
import operator

a = [1,3,5,6,2]
print(reduce(operator.add, a))
print(reduce(operator.mul, a))
print(reduce(operator.add, ["nugi"," ","saputra"])) 
```

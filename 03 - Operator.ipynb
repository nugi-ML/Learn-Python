Dalam pemrograman Python, Operator secara umum digunakan untuk melakukan operasi pada nilai dan variabel.
* Operator: Simbol khusus seperti $\text{-}$, $\text{+}$, $\text{*}$, $\text{/}$, dll.
* Operan: Nilai di mana operator diterapkan.

### Operator Arimatika

```python
# Variabel
a = 10
b = 20

# Penjumlahan
print(f'Hasil penjumlahan {a} + {b} = {a + b}')

# Pengurangan
print(f'Hasil pengurangan {b} - {a} = {b - a}')

# Perkalian
print(f'Hasil perkalian {a} * {b} = {a * b}')

# Pembagian
print(f'Hasil pembagian {b} / {a} = {b / a}')

# Modulus
print(f'Hasil modulus {b} % {a} = {b % a}')

# Eksponen
print(f'Hasil eksponen {a} ** 2 = {a ** 2}')

# Floor Division
print(f'Hasil floor division {b} // {a} = {b // a}')
```

### Operator Perbandingan

```python
c = 13
d = 33

# Lebih besar dari
print(f'Apakah {c} lebih besar dari {d}? {c > d}')

# Kurang dari
print(f'Apakah {c} kurang dari {d}? {c < d}')

# Sama dengan
print(f'Apakah {c} sama dengan {d}? {c == d}')

# Tidak sama dengan
print(f'Apakah {c} tidak sama dengan {d}? {c != d}')

# Lebih besar atau sama dengan
print(f'Apakah {c} lebih besar atau sama dengan {d}? {c >= d}') 

# Kurang dari atau sama dengan
print(f'Apakah {c} kurang dari atau sama dengan {d}? {c <= d}')
```

### Operator Logika

```python
a = True
b = False

# AND
print(f'Hasil dari {a} AND {b} adalah {a and b}')

# OR
print(f'Hasil dari {a} OR {b} adalah {a or b}')

# NOT
print(f'Hasil dari NOT {a} adalah {not a}')
```

### Operator Bitwise

```python
e = 10
f = 4

# Bitwise AND
print(f'Hasil Bitwise AND dari {e} & {f} adalah {e & f}')

# Bitwise OR
print(f'Hasil Bitwise OR dari {e} | {f} adalah {e | f}')

# Bitwise XOR
print(f'Hasil Bitwise XOR dari {e} ^ {f} adalah {e ^ f}')

# Bitwise NOT
print(f'Hasil Bitwise NOT dari ~{e} adalah {~e}')

# Bitwise Left Shift
print(f'Hasil Bitwise Left Shift dari {e} << 1 adalah {e << 1}')

# Bitwise Right Shift
print(f'Hasil Bitwise Right Shift dari {e} >> 1 adalah {e >> 1}')
```

### Operator Assignment

```python
g = 10
h = g

print(f'h = {h}')
print(f'b += g -> b = b + g = {h + g}')
print(f'b -= g -> b = b - g = {h - g}')
print(f'b *= g -> b = b * g = {h * g}')
print(f'b /= g -> b = b / g = {h / g}')
print(f'b %= g -> b = b % g = {h % g}')
```

### Operator Identitas

```python
a = 10
b = 20
c = a

print(f'Apakah c adalah a? {c is a}')
print(f'Apakah a adalah b? {a is b}')
print(f'Apakah c bukan b? {c is not b}')
```

Dalam Python, variabel digunakan untuk menyimpan data yang dapat dirujuk dan dimanipulasi selama eksekusi program. Variabel pada dasarnya adalah nama yang ditetapkan pada suatu nilai.

```python
x = 5
nama = 'Udin Kota baru'
print(x)
print(nama)
```

Aturan penamaan variabel:
* Nama variabel hanya boleh mengandung huruf, angka, dan garis bawah (_).
* Nama variabel tidak boleh diawali dengan angka.
* Nama variabel sensitif terhadap huruf besar/kecil (case-sensitive), seperti myVar dan myvar adalah berbeda.
* Hindari menggunakan kata kunci Python (Python keywords) seperti if, else, dan for sebagai nama variabel.

```python
# Benar
usia = 21
_warna = 'merah'
total_skor = 95.5

# Salah
# 2nd_name = 'Budi'
# my-variable = 10
# total skor = 50
```

```python
print(usia)
print(_warna)
print(total_skor)
```

```python
# Multi variabel
a, b, c = 1, 2, 3
print(a,b,c)
```

### Pengubahan Tipe (Type Casting) Variabel
* int(): Mengubah nilai yang kompatibel menjadi bilangan bulat (integer).
* float(): Mengubah nilai menjadi bilangan desimal (floating-point number).
* str(): Mengubah tipe data apa pun menjadi string (teks).

```python
angka = "10"
ubahAngka = int(angka)
print(type(angka))
print(type(ubahAngka))
```

```python
angka2 = 15.5
ubahAngka2 = int(angka2)
print(f'Angka adalah {angka2} dengan tipe {type(angka2)}')
print(f'Angka adalah {ubahAngka2} dengan tipe {type(ubahAngka2)}')
```

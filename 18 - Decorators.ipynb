### Decorators

* Decorator adalah fungsi di Python yang digunakan untuk memodifikasi atau menambahkan fungsionalitas ke fungsi lain tanpa mengubah kode asli fungsi tersebut.
* Decorator bekerja dengan cara membungkus fungsi target.

```python
def decorator(func):
    def wrapper(*args, **kwargs):
        # Tambahkan logika sebelum
        print('Sebelum fungsi dijalankan')
        hasil = func(*args, **kwargs)
        # Tambahkan logika sesudah
        print('Setelah fungsi dijalankan')
        return hasil
    return wrapper

@decorator
def halo(nama):
    print(f'Halo, {nama}!')

halo("Nugi")
```

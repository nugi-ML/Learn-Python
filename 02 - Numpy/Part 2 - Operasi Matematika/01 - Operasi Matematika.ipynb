```python
import numpy as np
```

Addtion of Arrays

```python
a = np.array([5,72,13,100])
b = np.array([2,5,10,30])

res = np.add(a,b)

print(res)
```

Subtraction of Arrays

```python
a = np.array([5,72,13,100])
b = np.array([2,5,10,30])

res = np.subtract(a,b)

print(res)
```

Multiplication of Arrays

```python
a = np.array([5,72,13,100])
b = np.array([2,5,10,30])

res = np.multiply(a,b)

print(res)
```

Division of Arrays

```python
a = np.array([5,72,13,100])
b = np.array([2,5,10,30])

res = np.divide(a,b)

print(res)
```

Exponentiation (Power)

```python
a = np.array([5,72,13,100])
b = np.array([2,5,10,30])

res = np.power(a,b)

print(res)
```

Modulus Operation

```python
a = np.array([5,72,13,100])
b = np.array([2,5,10,30])

res = np.mod(a,b)

print(res)
```

**Fungsi Trigonometri**

```python
import numpy as np

angles = np.array([0,30,45,60,90])
rad = np.deg2rad(angles) # Konversi degres ke radian

sin_val = np.sin(rad)
print("Nilai sin: ",sin_val)

inv_sin = np.rad2deg(np.arcsin(sin_val))
print("Inversi sin (degrees): ",inv_sin)

sinh_val = np.sinh(rad)
print("Sin Hiperbola: ",sinh_val)
```

**Fungsi Statistik**

```python
berat = np.array([50.7, 52.5, 50, 58, 55.63, 73.25, 49.5, 45])

print(f"Min dan Max: {np.amin(berat)}, {np.amax(berat)}")
print(f"Range: {np.ptp(berat)}")
print(f"70th Percentile: {np.percentile(berat, 70)}")
print(f"Mean: {np.mean(berat)}")
print(f"Median: {np.median(berat)}")
print(f"Std Dev: {np.std(berat)}")
print(f"Variance: {np.var(berat)}")
print(f"Average: {np.average(berat)}")
```

**Fungsi Bitwise**

```python
genap  = np.array([2,4,6,8,10,12,14])
ganjil = np.array([1,3,5,7,9,11,13])

# Bitwise AND, OR, XOR
print(f"AND: {np.bitwise_and(genap,ganjil)}")
print(f"OR: {np.bitwise_or(genap,ganjil)}")
print(f"XOR: {np.bitwise_xor(genap,ganjil)}")

# Bitwise NOT
print(f"Invert: {np.invert(genap)}")

# Bit shifts
print(f"Left shifts: {np.left_shift(genap,1)}")
print(f"Right shifts: {np.right_shift(genap,1)}")
```

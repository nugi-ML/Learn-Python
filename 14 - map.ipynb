### map()

* map() adalah fungsi bawaan Python yang digunakan untuk menerapkan sebuah fungsi ke setiap elemen dari iterable (seperti list, tuple, dll).
* map(function, iterable)
* function → fungsi yang akan diterapkan (bisa def biasa atau lambda).
* iterable → kumpulan data (list, tuple, dll).



```python
s = ['1', '2', '3', '4'] 
res = map(int, s)
print(list(res))
```

```python
def double(val):
    return val * 2

a = [1,2,3,4]
res = list(map(double, a))
print(res)
```

```python
a = [1,2,3,4]
res = list(map(lambda x: x * 2, a))
print(res)
```

```python
a = [1,2,3]
b = [4,5,6]
res = map(lambda x, y: x + y, a, b)
print(list(res))
```

konversi string ke uppercase

```python
fruits = ['apple', 'banana', 'cherry']
res = map(str.upper, fruits)
print(list(res))
```

Mengambil huruf pertama

```python
words = ['apple', 'banana', 'cherry']
res = map(lambda s: s[0], words)
print(list(res))
```

Menghapus space dari string

```python
s = ['  hello  ', '  world', 'python  ']
res = map(str.strip, s)
print(list(res))
```

Konversi celsius ke fahrenheit

```python
celsius = [0, 20, 37, 100]
fahrenheit = map(lambda c: (c * 9/5) + 32, celsius)
print(list(fahrenheit))
```

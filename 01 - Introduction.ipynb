```python
# Ini adalah komentar. Tidak akan dieksekusi oleh interpreter Python.
print("Hallo, World!")  # Ini juga komentar setelah kode.
```

* print() adalah fungsi bawaan (built-in) Python yang menginstruksikan komputer untuk menampilkan teks di layar.  
* "Hello, World!" adalah sebuah string, yang merupakan urutan dari teks. Dalam Python, string diapit oleh tanda kutip (bisa tunggal ' atau ganda ").
* Apa pun setelah simbol # adalah komentar. Python mengabaikan komentar, tetapi komentar berguna untuk menjelaskan kode kepada pembaca.

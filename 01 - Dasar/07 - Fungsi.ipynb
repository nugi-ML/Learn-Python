```python
def evenOdd(x):
    if x % 2 == 0:
        return "Genap"
    else:
        return "Ganjil"
    
print(evenOdd(10))  # Output: Genap
print(evenOdd(7))   # Output: Ganjil
```

#### Jenis-jenis Argumen Fungsi

Default Arguments

```python
def myFun(x, y=50):
    print("x:", x)
    print("y:", y)

myFun(10) 
```

Keyword Arguments

```python
def student(fname, lname):
    print(fname, lname)

student("John", "Doe")
```

Positional Arguments

```python
def nameAge(name, age):
    print("Name:", name)
    print("Age:", age)

print('Case-1:')
nameAge("Alice", 30)

print('\nCase-2:')
nameAge(age=25, name="Bob")
```

Arbitrary Arguments

Dalam Argumen Kata Kunci Arbitrari (Arbitrary Keyword Arguments) Python, *args dan **kwargs dapat meneruskan sejumlah variabel argumen ke suatu fungsi menggunakan simbol khusus. Ada dua simbol khusus:  
* *args (Argumen Non-Kata Kunci)
* **kwargs (Argumen Kata Kunci)

```python
def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

myFun('Hey', 'Welcome', first='Udin', mid='adalah', last='manusia')
```

Anonymous Functions

```python
def cube(x): return x*x*x   # without lambda
cube_l = lambda x : x*x*x  # with lambda

print(cube(7))
print(cube_l(7))
```

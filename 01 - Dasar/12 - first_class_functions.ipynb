Menetapkan Fungsi ke Variabel

```python
def msg(name):
    return f"Hello, {name}"

f = msg

print(f('Nugi'))
```

Meneruskan Fungsi sebagai Argumen

```python
def msg(name):
    return f"Hello, {name}"

def fun1(fun2, name):
    return fun2(name)

print(fun1(msg, "Udin"))
```

Mengembalikan Fungsi dari Fungsi Lain

```python
def fun1(msg):
    def fun2():
        return f"Message: {msg}"
    return fun2

func = fun1("Hai Semuanya")
print(func())
```

Menyimpan Fungsi dalam Struktur Data

```python
def add(x,y):
    return x + y

def subtract(x,y):
    return x - y

d = {
    "add": add,
    "subtract": subtract
}

print(d["add"](5, 3))
print(d["subtract"](5, 3))
```

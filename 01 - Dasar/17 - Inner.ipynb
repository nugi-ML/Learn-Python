### Inner Functions

* Inner function adalah fungsi yang didefinisikan di dalam fungsi lain.
* Kadang disebut juga nested function.
* Tujuannya biasanya untuk membatasi ruang lingkup fungsi agar hanya bisa dipakai di dalam fungsi induknya.

```python
def fun1(msg):
    def fun2():
        print(msg)
    fun2()
fun1("Python")
```

Modifikasi variavel menggunakan nonlocal

```python
def fun1():
    a = 45
    def fun2():
        nonlocal a
        a = 52
        print(a)
    fun2()
    print(a)
fun1()
```

Closure di fungsi inner

```python
def fun1(a):
    def fun2():
        print(a)
    return fun2

closure_func = fun1("Hello, closure!")
closure_func()
```

### filter()

filter() adalah fungsi bawaan Python yang digunakan untuk menyaring elemen-elemen dari iterable (list, tuple, dll) berdasarkan kondisi tertentu.

filter(function, iterable)

* function → fungsi yang menerima satu argumen dan mengembalikan True atau False.
* iterable → kumpulan data (list, tuple, dll).
* Hasilnya berupa filter object (iterator), biasanya diubah ke list atau tuple.

```python
def even(n):
    return n % 2 == 0

a = [1,2,3,4,5,6]
b = filter(even, a)
print(list(b))
```

```python
a = [1,2,3,4,5,6]
b = filter(lambda x: x % 2 == 0, a)
print(list(b))
```

```python
a = [1,2,3,4,5,6]
b = filter(lambda x: x % 2 == 0, a)
c = map(lambda x: x * 2, b)
print(list(c))
```

```python
a = ["apple", "banana", "cherry", "kiwi", "grape"]
b = filter(lambda w: len(w) > 5, a)
print(list(b))
```

```python
L = ['apple', '', None, 'banana', 0, 'cherry']
a = filter(None, L)
print(list(a))
```

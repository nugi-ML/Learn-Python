### Tipe data numerik

```python
a = 5
print(type(a))

b = 3.14
print(type(b))

c = 2 + 4j
print(type(c))
```

### Tipe data sequence

Tipe data string

```python
s = "Hello, World!"
print(s)

print(type(s))

# Akses string dengan indeks
print(s[0])
print(s[7:12])
```

Tipe data list

```python
# List kosong
a = []
print(a)

# List dengan beberapa elemen
b = [1, 2, 3, 4, 5]
print(b)

# List dengan berbagai tipe data
c = [1, "dua", 3.0, True]
print(c)
```

```python
# akses elemen list dengan indeks
print(b)
print(b[0])    # elemen pertama
print(b[1:4])  # elemen kedua sampai keempat
print('====================')
print(c)
print(c[-1])  # elemen terakhir
print(c[-3])  # elemen ketiga dari belakang
```

Tipe data tuple

```python
# Tuple kosong
tup1 = ()

tup2 = ('satu', 'dua', 'tiga')
print(tup2)
```

Tipe data boolean

```python
print(type(True))
print(type(False))
```

```python
if 1:
    print("1 adalah truthy")

if not 0:
    print("0 adalah falsy")
```

Tipe data set

```python
# Set kosong
set1 = set()

set2 = set('Hello World')
print('Set dari string:', set2)

set3 = set(['Hello', 'World'])
print('Set dari list:', set3)
```

Tipe data dictionary

```python
# Dictionary kosong
dict1 = {}

dict2 = {'nama': 'Alice', 'umur': 25}
print(dict2)
```

```python
# Akses nilai dictionary dengan kunci
print('Nama:', dict2['nama'])
print('Umur:', dict2['umur'])
```

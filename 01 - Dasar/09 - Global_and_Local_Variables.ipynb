### Local Variabel

```python
def greet():
    msg = "Belajar python sangat menyenangkan"
    print(msg)

greet()
```

```python
def greet():
    msg = 'Hallo'
    print('Di dalam fungsi : ',msg)

greet()
print('Di luar fungsi : ',msg)
```

### Global Variabels

```python
msg = "Python sangat menyenangkan"

def display():
    print("Di dalam fungsi : ",msg)

display()
print("Di luar fungsi : ",msg)
```

### Use of Local and Global variables

```python
def fun():
    s = "Me too"
    print(s)

s = "Saya sangat suka bahasa python"
fun()
print(s)
```

### Modifying Global Variables Inside a Function

Without global (causes error)

```python
def fun():
    s += "KEREN"
    print(s)

s = "Python bagus sekali"
fun()
```

With global (works correctly)

```python
s = "Python menyenangkan"

def fun():
    global s
    s += " Keren"
    print(s)
    s = "Python sangat mudah"
    print(s)

fun()
print(s)
```

Global vs Local with Same Name

```python
a = 1

def f():
    print("f():", a) # menggunakan global a

def g():
    a = 2 # lokal menimpa global a
    print("g():", a)

def h():
    global a
    a = 3 # Mengubah global a
    print("h():", a)

print("global:", a)
f()
print("global:", a)
g()
print("global:", a)
h()
print("global:", a)
```

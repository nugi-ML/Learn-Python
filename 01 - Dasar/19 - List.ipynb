Adding Elements into List

append(): Menambahkan sebuah elemen di akhir daftar.

extend(): Menambahkan beberapa elemen ke akhir daftar.

insert(): Menambahkan sebuah elemen pada posisi tertentu.

clear(): Menghapus semua elemen dari daftar.

```python
a = []

a.append(10)
print('After append(10):', a)

a.insert(0,5)
print("After insert(0,5):", a)

a.extend([15,20,25])
print("After extend([15,20,25]):", a)

a.clear()
print('after clear():', a)
```

Updating Elements into List

```python
a = [10,20,30,40,50]
a[1] = 25
print(a)
```

Removing Elements from List

remove(): Menghapus kemunculan pertama dari sebuah elemen.

pop(): Menghapus elemen pada indeks tertentu, atau elemen terakhir jika indeks tidak ditentukan.

del statement: Menghapus elemen pada indeks yang ditentukan.

```python
a = [10,20,30,40,50]

a.remove(30)
print('After remove(30):', a)

popped_val = a.pop(1)
print('Popped element:', popped_val)
print('After pop(1):', a)

del a[0]
print('Adter del a[0]:', a)
```

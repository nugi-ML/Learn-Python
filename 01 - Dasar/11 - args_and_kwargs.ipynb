### Non-Keyword Arguments (*args)

* Digunakan untuk menangkap argumen positional (tanpa nama).
* Hasilnya berupa tuple.
* Cocok ketika kita tidak tahu berapa banyak argumen yang akan diberikan.

```python
def myFun(*args):
    for arg in args:
        print(arg)

myFun('Hello', 'My', 'Name is', 'Nugie Saputra')
```

```python
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(multiply(2,3,6))
```

### Keyword Arguments (**kwargs)

* Digunakan untuk menangkap argumen keyword (berpasangan nama=nilai)
* Hasilnya berupa dictionary
* Cocok untuk fungsi yang menerima parameter opsional dengan nama.

```python
def fun(**kwargs):
    for key, val in kwargs.items():
        print(f"{key}: {val}")

fun(name="Nugi", age="21", country="Indonesia")
```

```python
def introduce(**kwargs):
    details = []
    for k, v in kwargs.items():
        details.append(k + ": " + str(v))
    return ", ".join(details)

print(introduce(name="Nugi", age=21, city="Bandung"))
```

### Using both *args and **kwargs

```python
def student_info(*args, **kwargs):
    print(f"Subjects : {args}")
    print(f"Details : {kwargs}")

student_info("Math", "Physics", "English", Name="Nugi", age=25, city="Jakarta")
```

For Loop

```python
n = 4
for i in range(n):
    print("Perulangan ke-", i+1)
```

```python
li = [10, 20, 30, 40]
for item in li:
    print("Item:", item)
```

While Loop

```python
cnt = 0
while cnt < 5:
    print("Hitung:", cnt)
    cnt += 1
```

```python
# while Loop tanpa batas
while True:
    print("Hallo dunia!")
```

Loops Bersarang

```python
for i in range(1,5):
    for j in range(i):
        print(i, end=' ')
    print()
```

Continue Statement  
Tidak mengeksikusi suatu index

```python
for latter in 'indonesia':
    if latter == 'o':
        continue
    print(latter)
```

Break Statement  
Stop perulangan

```python
for latter in 'indonesia':
    if latter == 'o':
        break
    print(latter)
```

Pass Statement  
Tidak mempengaruhi apa-apa

```python
for latter in 'indonesia':
    if latter == 'o':
        pass
    print(latter)
```

```python
d1 = {1: 'bahasa', 2: 'pemograman', 3: 'Python'}
print(d1)

d2 = dict(a = 'bahasa', b = 'pemograman', c = 'Python')
print(d2)
```

Accessing Dictionary Items

```python
d = {"nama": "nugi", "usia": 21}

# Access using key
print(d["nama"])

# Acces using get
print(d.get("nama"))
```

Adding and Updating Dictionary Items

```python
d = {"nama": "nugi", "usia": 21}

# Menambahakan
d["kota"] = "Jakarta"

# Update
d["nama"] = "Udin"

print(d)
```

Removing Dictionary Items

```python
d = {"nama": "nugi", "usia": 21, "kota": "bandung"}

# Using del
del d["usia"]
print(d)

# Using pop
val = d.pop("nama")
print(val)

# Using popitem
key, val = d.popitem()
print(f"key:  {key}, value: {val}")

# Using clear
d.clear()
print(d)
```

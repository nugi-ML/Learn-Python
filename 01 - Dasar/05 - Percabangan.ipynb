if statement

```python
usia = 21
if usia >= 18:
    print("Anda sudah dewasa.")
```

```python
if usia >= 18: print("Anda sudah dewasa.") 
```

if else statement

```python
usia = 16
if usia >= 18:
    print("Anda sudah dewasa.")
else :
    print("Anda masih di bawah umur.")
```

```python
res = "Anda sudah dewasa." if usia >= 18 else "Anda masih di bawah umur."
print(res)
```

elif statement

```python
usia = 25

if usia <= 12:
    print("Anak-anak")
elif usia <= 18:
    print("Remaja")
elif usia <= 35:
    print("Dewasa")
else:
    print("Lansia")
```

if..else bersarang

```python
total = 100000
is_member = True

if total >= 80000:
    if is_member:
        print("Diskon 30%")
    else:
        print("Diskon 10%")
else:
    print("Tidak ada diskon")
```

Recursion dalam Python adalah teknik pemrograman di mana sebuah fungsi memanggil dirinya sendiri untuk menyelesaikan masalah dengan cara memecahnya menjadi submasalah yang lebih kecil.

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:       # Recursive case
        return n * factorial(n-1)

factorial(5)
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci(10)
```

### Types of Recursion in Python
* Tail Recursion: Pemanggilan rekursif adalah hal terakhir yang dilakukan fungsi, sehingga tidak ada pekerjaan lain setelah fungsi itu mengembalikan hasil. Beberapa bahasa pemrograman dapat mengoptimalkan hal ini agar bekerja seperti sebuah loop, sehingga menghemat memori.
* Non-Tail Recursion: Fungsi masih melakukan pekerjaan tambahan setelah pemanggilan rekursif mengembalikan hasil, sehingga tidak bisa dioptimalkan menjadi sebuah loop.

```python
def tail_fact(n, acc=1):
    if n == 0:
        return acc
    else:
        return tail_fact(n-1, acc*n)

def nontail_fact(n):
    if n == 1:
        return 1
    else:
        return n * nontail_fact(n-1)

print(tail_fact(5))
print(nontail_fact(5))
```

```python
set1 = {1,2,3,4,5}
print(set1)
```

```python
set1 = set(['python', 'java', 'C++'])
print(set1)
```

Adding Elements to a Set in Python

```python
set1 = {1,2,3}

set1.add(4)

set1.update([5,6])

print(set1)
```

Accessing a Set in Python

```python
set1 = set(['python', 'java', 'C++'])

print("python" in set1)
```

Removing Elements from the Set in Python

```python
set1 = {1,2,3,4,5}
set1.remove(3)
print(set1)
```

```python
set1.discard(4)
print(set1)
```

```python
set1 = {1,2,3,4,5}
val = set1.pop()
print(val)
print(set1)
```

```python
set1 = {1,2,3,4,5}
set1.clear()
print(set1)
```
